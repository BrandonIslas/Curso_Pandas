# Graficos con Pandas

```python
import pandas as pd
import numpy as np
```

```python
dataframe= pd.DataFrame(np.random.randint(200,size=(50,4)),columns=['a','b','c','d'])
dataframe
```

```python
dataframe['a'].hist()
```

```python
dataframe['b'].hist(bins=30)
```

```python
dataframe.plot.area()
```

```python
dataframe.plot.bar()
```

```python
dataframe.plot.bar(stacked=True)
```

```python
dataframe.plot.scatter(x='a',y='b',c='c',cmap='coolwarm')
```

```python
dataframe1= pd.DataFrame(np.random.randint(200,size=(50,4)),columns=['a','b','c','d'])
dataframe1.plot.box()
```

```python
dataframe.plot.hexbin(x='a',y='b',gridsize=12)
```

```python
dataframe.plot.kde()
```

```python
dataframe.plot.density()
```

```python
dataframe2= pd.DataFrame(np.random.randint(200,size=(50,4)),columns=['a','b','c','d'])
dataframe2.plot.pie(x='a',y='b',figsize=(100,100))
```
