## Pairgrid y FaceGrid

```python
flores= sns.load_dataset('iris')
flores.head()
```

```python
sns.pairplot(flores)
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter)
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.distplot)
graficos.map_upper(plt.scatter)
graficos.map_lower(sns.kdeplot)
```
